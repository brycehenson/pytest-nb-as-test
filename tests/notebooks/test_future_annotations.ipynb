```python
from __future__ import annotations


def repeat_text(text: str, count: int) -> list[str]:
    """Repeat text values.

    Args:
        text: Text to repeat.
        count: Number of repetitions.

    Returns:
        Repeated text values.

    Example:
        repeat_text("hi", 2)
    """
    return [text] * count
```

```python
result = repeat_text("hi", 3)
result
```
