```python
from __future__ import annotations


def stringify(value: int) -> str:
    """Convert integers to strings.

    Args:
        value: Integer value to convert.

    Returns:
        Converted string value.

    Example:
        stringify(3)
    """
    return str(value)
```

```python
assert stringify(3) == "3"
```
